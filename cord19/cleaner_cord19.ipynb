##  Limpeza dos dados do dataset Cord19

   Esta atividade tem como objetivo fazer a coleta e leitura  do corpus fornecido pela plataforma do Cord-19, bem como realizar a limpeza dos dados, removendo colunas que não serão utilizadas e valores nulos. 


```python
import pandas as pd
```

Criando DataFrame pandas

```python
filepath = "/content/drive/MyDrive/Colab Notebooks/cord-19/metadata.csv"
df = pd.read_csv(filepath)
```

```python
display(df.tail(3))
```

```python
print(df.columns)
```

```python
print(df.shape)
```

Excluindo colunas que não serão utilizadas para análise

```python
df.drop(['cord_uid','sha', 'source_x','authors', 
         'journal','doi', 'pmcid', 'pubmed_id','license', 
         'mag_id','who_covidence_id', 'arxiv_id', 'pdf_json_files',
         'pmc_json_files','s2_id'], 
        axis='columns', inplace=True)
```

```python
print(df.shape)
```

```python
print(df.columns)
```

Checando e removendo *missing values* do dataframe

```python
df.isna().sum()
```

```python
df.dropna(subset = ['title','publish_time','abstract','url'], inplace=True)
```

```python
df.isna().sum()
```

```python
print(df.shape)
```

Removendo artigos antigos (publicados antes de 2020)

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

```python
df.drop(df.loc[df['publish_time'] <= '2019-12-31'].index, inplace=True)
```

Apenas dados completos e com as colunas selecionadas, gerou um novo dataframe <Original=(529651, 19);Processado=(140265, 7)>.

```python
print(df.shape)
```

```python
display(df.tail(3))
```

Gerando um arquivo .txt para uso posterior

```python
df.to_csv('dataset_cord19.txt', sep='\t', index=False)
```
